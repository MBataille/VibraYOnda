**Ejemplo 1: Masa con un resorte**

Es evidente que al encontrarse con el típico sistema de una masa $m$ conectada a un resorte de constante $k$, se obtiene:
\begin{align}
\ddot{x} &= -\omega^2x \\
\omega^2 &= \frac{k}{m}
\end{align}

**Ejemplo 2: Masa en mínimo de potencial**

Al trabajar con un potencial $U$ que admite un mínimo en algún punto $x_{min}$, nos encontramos nuevamente con un oscilador armónico. En este caso el análisis matemático es distinto. Primero realizamos una expansión de taylor de segundo grado en el punto de equilibrio. Recordemos que cerca del mínimo la pendiente es cero, por lo tanto $U'(x_{min}) = 0$, y por construcción $U_0=0$.

\begin{align}
U(x) &= U_0 + U'(x_{min})x + \frac12 U''(x_{min})x^2 \\
    &= \frac12 U''(x_{min})x^2 \\
    \\
F &= -\frac{d}{dx}U \\
 &= U''(x_{min})x \\
 \\
 m\ddot{x} &= -U''(x_{min})x \\
\omega^2 &= \frac{U''(x_{min})}{m} \\
\end{align}


Ahora que podemos reconocer la ecuación diferencial del oscilador armónico, es necesario poder resolverla!

**Solución oscilador armónico**

Son muchas las formas posibles de resolver la ecuación diferencial $$\ddot{x} = -\omega^2x $$

En este caso veremos la solución de Landau:

\begin{align}
z &= x + i\frac{\dot{x}}{\omega} \\ \\
\dot{z} &= \dot{x} + i\frac{\ddot{x}}{\omega} \\
&= -i\omega z
\end{align}

Se deduce entonces que $$z = z_0e^{-i\omega t}$$
Notemos que se puede escribir $z_0$ como:

$$z_0 = Ae^{i\phi}$$
Donde $A$ es la *amplitud* del movimiento y $\phi$ es la *fase*

Finalmente, podemos escribir
\begin{align}
x &= A\cos(\omega t + \phi) \\
\dot{x} &= A\omega\sin(\omega t + \phi)
\end{align}

**Notación compleja **

Otro beneficio de adoptar la notación de la solución de Landau, es la propiedad siguiente:

$$zz^* = Ae^{i\phi-\omega t} \times Ae^{-i\phi + \omega t} = A^2$$

Donde $z^{*}$ representa el conjugado de $z$.
Por otro lado, $$zz^{*} = |z|^2 = x^2+ \frac{\dot{x}^2}{\omega^2} = A^2 $$

Esta última ecuación se puede representar como una elipse en un espacio de fases, como en el gráfico siguiente. Donde cada punto representa una condición inicial posible.



```python
A = 5
x = np.linspace(-A,A,101)
y1 = np.sqrt(A**2-x**2)
y2 = -np.sqrt(A**2-x**2)

plt.plot(x,y1,'b-')
plt.plot(x,y2,'b-')

plt.xlabel('x')
plt.ylabel('v')
plt.show()
```


### Disipación

**Roce viscoso**

En principio, el roce viscoso se puede escribir como:

$$ F = -\eta v$$

Luego, la ecuación del oscilador armónico amortiguado para una masa conectada a un resorte queda de la forma

$$m\ddot{x} + \eta\dot{x} + \kappa x = 0$$

**Ejemplo: Resistencia**

En un circuito RLC, la suma de los voltajes de cada componente es nula ya que es un circuito cerrado. Luego,



$$L\dot{I} + RI + \frac{Q}{C} = 0$$ 

Considerando $I=\frac{dQ}{dt}$,

$$L\ddot{Q} + R\dot{Q} + \frac{Q}{C} = 0$$

**Osilador armónico amortiguado**

Escribiremos la ecuación del oscilador armónico como:

$$\ddot{x} + 2\gamma \dot{x} + \omega_0^2 x = 0$$

Multiplicamos por $\dot{x}$:

$$\dot{x}\ddot{x} + \dot{x}\omega_0^2 x = -2\gamma\dot{x}^2$$

Podemos ver que el término a la izquierda equivale a la derivada de la energía del sistema:

$$\frac{d}{dt}\left(\frac{\dot{x}^2}{2} + \omega_0^2 \frac{x^2}{2}\right) = -2\gamma\dot{x}^2$$

Notemos entonces que el término de la izquierda (que antes era igual a $A^2$), ahora deja de ser constante y va disminuyendo en el tiempo.

Lo que antes era una elipse en el espacio de fases, ahora se transforma en un espiral en el espacio de fases.

Notemos entonces que dependiendo de si $\gamma$ es muy grande comparado a $\omega$, entonces el sistema no alcanza a osiclar y decae muy rápidamente. En el caso contrario, el sistema oscila varias veces y su amplitud disminuye lentamente.

Se define entonces el llamado factor de calidad del oscilador:

$$Q = \frac{\omega_0}{\gamma}$$

Dicho factor de calidad es mayor para sistemas con menor disipación. 

**Solución de la ecuación**

Asumiremos $x$ de la forma $e^{\lambda t}$

Luego, queda el polinomio característico:

$$\lambda^2 + 2\gamma \lambda + \omega_0^2 = 0$$

De donde obtenemos la solución:

$$\lambda = -\gamma \pm \sqrt{\gamma^2-\omega_0^2}$$

O, en el caso de no existir solución real:

$$\lambda = -\gamma + i\sqrt{\omega_0^2 - \gamma^2} $$


De acá podemos identificar una variedad de casos:

$\omega_0 <\gamma$ es el caso sobreamortiguado, no hay oscilación.

$\omega_0 = \gamma$ es el caso de amortiguamiento crítico.

El caso imaginario corresponde a una oscilación (ya que aparece la exponencial imaginaria).

Finalmente:

$$ x = Re\left(z_0 e^{-\gamma t} e^{\pm i\sqrt{\omega_0^2-\gamma^2}}\right) $$

De donde podemos notar que la frecuencia de oscilación queda $\omega =\sqrt{\omega_0^2- \gamma^2}$


### Oscilador Forzado y Resonancia

La resonancia es la respuesta de un oscilador a un estímulo externo. El caso típico siendo el de una masa con un resorte colgando de un techo que oscila a cierta frecuencia.


** Ejemplo: Spin Nuclear**

Tomemos el caso de un sólo protón. El protón tiene una característica llamada el spin, es decir, la capacidad de girar en torno a sí mismo. Al estar cargado, aparece una carga en movimiento que se traduce en un campo magnético. El protón tiene entonces un cierto momento magnético $\vec{m}$.

Así el núcleo, compuesto por protones y neutrones, tiene también un momento magnético $\vec{m}$. Al estar inmerso en un campo magnético externo, se genera un torque sobre el núcleo tal que, $$\frac{d\vec{m}}{dt} = \gamma\vec{B}\times\vec{m}$$ 

***Tarea: Demostrar la relación anterior***

Además, se cumple que el módulo de $\vec{m}$ es constante ya que

$$\frac{d}{dt}(\vec{m}^2) = 2\vec{m}\cdot\frac{d\vec{m}}{dt} = 0$$

Diremos que $\vec{m} = m_0\hat{\Omega}$ y supondremos que $\hat{\Omega} \approx \hat{z}$.

Por lo tanto $\hat{\Omega} = (\Omega_x,\Omega_y,1-\frac12(\Omega_x^2+ \Omega_y^2)$.

\begin{align}
\frac{dm_x}{dt} &= m_0 \frac{d\Omega_x}{dt} \\
 &= \gamma B \hat{z}\times\vec{m} \\
 &= -m_0\gamma B \Omega_y  \\
\end{align}

Por otro lado,

\begin{align}
\frac{d\Omega_y}{dt} =+\gamma B\Omega_x
\end{align}

Derivamos y encontramos la ecuación del movimiento armónico simple,

\begin{align}
\frac{d^2\Omega_x}{dt^2} &= -\gamma B\frac{d\Omega_y}{dt} \\
    &= -\gamma B \gamma B\Omega_x \\
    &= -(\gamma B)^2 \Omega_x
\end{align}

Queda entonces que $\omega = \gamma B$, donde $\gamma$ cambia para cada átomo.
En el caso de los exámenes de resonancia magnética, se aplica un campo magnético como el siguiente:

$$\vec{B} = B\hat{z} + B_x\cos{\omega t}\hat{x} $$

***Demostrar que esto lleva a la ecuación de resonancia***

**Ecuación de resonancia:**

Intentaremos resolver la ecuación de resonancia.

\begin{align}
\ddot{x} + \gamma\dot{x} + \omega_0^2 x &= F\cos(\omega t) \\ \\
\ddot{z} + \gamma \dot{z} + \omega_0^2 z &= Fe^{i\omega t} \\
\text{con }z &= z_0 e^{i\omega t} \\
\text{queda entonces }(\omega_0^2 - \omega^2 + i\gamma \omega)z_0 &= F \\
Ae^{i\delta} z_0 &= F \text{ término de la izquierda en notación exponencial} \\ \\
z_0 = \frac{F}{A} e^{-i\delta} \\
\end{align}

Queda finalmente que
\begin{align}
A^2 &= (\omega_0^2-\omega^2)^2+\gamma^2\omega^2 \\
\tan \delta &= \frac{\gamma \omega}{\omega_0^2-\omega^2} \\
\end{align}

De donde,
$$x(t) = \frac{F}{\sqrt{(\omega_0^2-\omega^2)^2+\gamma^2\omega^2}} \cos(\omega t-\delta) $$


**Frecuencia alta**

En este caso, $\omega >> \omega_0$. Así queda que

$$x(t) = -\frac{F}{\omega^2} \cos(\omega t)$$

Por lo que la respuesta ($x(t)$) está desfasada del forzamiento ($F\cos(\omega t)$).

**Frecuencia baja**

Para $\omega << \omega_0$, queda que

$$x(t) = \frac{F}{\omega_0^2} \cos(\omega t)$$

Así, la respuesta y el forzamiento están en fase aunque la amplitud de la respuesta es mucho menor que la del forzamiento.

**Resonancia**

Buscamos la frecuencia del forzamiento para la cual la amplitud de la respuesta es mayor. Derivando, encontramos que,

\begin{align}
2(\omega_0^2 - \omega^2)(-2\omega) + 2\gamma^2\omega &= 0 \\ 
\omega^2 &= \omega_0^2-\frac{\gamma^2}{2} \\
\end{align}

Notemos que la frecuencia de resonancia es ligeramente menor que la frecuencia natural del sistema. Además, evaluando la tangente obtenemos que $\delta \approx \frac{\pi}{2}$.


** Trabajo de la fuerza externa **

Potencia se escribe como

\begin{align}
P &= Fv \\
 &= -F\cos(\omega t) \frac{F}{A}\omega\sin(\omega t-\delta) \\
 &= \frac{F^2\omega_0}{Q}\frac{1}{(\frac{\omega_0}{\omega}-\frac{\omega}{\omega_0})^2 + \frac{1}{Q^2}}
\end{align}


** Potencia promedio por ciclo **

$$P = \frac{1}{T} \int_0^T{P(t)dt} $$

Recordemos que como trabajamos con movimientos armónicos, la integral de arriba tendrá senos y/o cosenos, por lo tanto es importante recordar propiedades de la integral.

\begin{align}
\int_0^{2\pi}  \cos(nx)dx &= 0 \\
\int_0^{2\pi} \sin(nx)dx &= 0 \\
\int_0^{2\pi} \cos^2(x)dx &= \pi \\
\int_0^{2\pi} \cos(x)\sin(x)dx &= 0 \\
\end{align}

La integral de la potencia queda entonces de la forma

\begin{align}
P &= \frac{\omega}{2\pi}\int_0^{2\pi/\omega} \cos(\omega t)\sin(\omega t-\delta) \\
&= \frac{\omega}{2\pi} \int_0^{2\pi} \cos\tau\sin(tau-\delta)d\tau \\
&=\frac{\omega}{2\pi} \int_0^{2\pi} \cos\tau\left(\sin\tau\cos\delta-\sin\delta\cos\tau\right)d\tau \\
&= -\frac12\sin\delta \\
\end{align}

Agregamos las constantes que multiplican y obtenemos,

$$ P = \frac{\omega}{2} \frac{F^2\sin\delta}{\sqrt{(\omega^2 -\omega_0^2) + (\gamma\omega)^2}} $$

## Unidad 2: Mecánica Lagrangeana

### Coordenadas Generalizadas.

Se tienen $N$ partículas con coordenadas cartesianas por ejemplo. Llamaremos $q_i$ las coordenadas generalizadas, es decir, las coordenadas que cumplen todas las restricciones del sistema. A partir de las coordenadas generalizadas podemos encontrar las coordenadas (cartesianas) del sistema. Así las coordenadas del sistema quedan como funciones de las coordenadas generalizadas.

###  Principio de mínima Acción.

Escribimos la energía cinética $K$ y la energía potencial $V$ en función de las coordenadas generalizadas. La acción se define como

$$S = \int_{t_1}^{t_2}(K-V)dt$$

Y la acción es mínima para la trayectoria clásica.

Notemos que la acción depende de $x(t)$ y $\dot{x}(t)$. Se le llama entonces funcional ya que su dominio son funciones.

Definimos también el Lagrangeano $\mathcal{L}$ tal que

$$\mathcal{L} = K - V$$


### Ecuacion de Euler-Lagrange

Supongamos $q$ es el mínimo de la acción. Como se trata de un mínimo, la derivada es nula, por lo tanto
$$ S(q+\delta q) \approx S(q) + O(\delta q^2)$$

Luego evaluamos la acción en $q+\delta q$

\begin{align}
S [q+\delta q, \dot{q} + \delta\dot{q} ] &= \int \mathcal{L}(q+\delta q, \dot{q}+\delta\dot{q},t)dt \\
&= \int \left(\mathcal{L}(q,\dot{q},t) + \frac{\partial \mathcal{L}}{\partial q}\delta q + \frac{\partial \mathcal{L}}{\partial \dot{q}}\delta \dot{q} + O(\delta q^2)\right)dt \\
\end{align}

Tenemos entonces que para cumlpir nuestra hipótesis, se tiene que cumplir que para todo $\delta q$

\begin{align}
\int \left(\frac{\partial \mathcal{L}}{\partial q}\delta q + \frac{\partial \mathcal{L}}{\partial \dot{q}}\delta \dot{q}\right)dt &= 0 \\
\int \delta q\left(\frac{\partial \mathcal{L}}{\partial q} - \frac{d}{dt}\left(\frac{\partial \mathcal{L}}{\partial \dot{q}}\right)\right)dt &= 0 \\
\end{align}

De donde deducimos que,

$$\frac{\partial \mathcal{L}}{\partial q} - \frac{d}{dt}\left(\frac{\partial \mathcal{L}}{\partial \dot{q}}\right) = 0$$

## Unidad 3: Modos Normales


***Langrangeano de un péndulo doble***

El langrangeano de un péndulo doble puede quedar expresado como 

$$M\ddot{\phi} = -K\phi$$

Donde $ M = m\ell^2\begin{pmatrix}
2 & 1 \\ 1 & 1
\end{pmatrix} $, $K = mg\ell\begin{pmatrix}
2 & 0 \\ 0 & 1
\end{pmatrix}$ y $\phi = \begin{pmatrix}
\phi_1 \\ \phi_2 \end{pmatrix} $

Buscamos los modos normales, por lo que asumimos solución de la forma

$$\phi = \begin{pmatrix} \phi_1^0 e^{i\omega t} \\ \phi_2^0 e^{i\omega t} \end{pmatrix} $$

Luego, reemplazando en la ecuación de movimiento obtenemos,

$$(K-\omega^2 M) \phi^0 = 0$$

Forzamos que el determinante del término de la izquiera sea cero (de lo contrario sería invertible y queda una solución trivial de $\phi^0$). Al hacer esto obtenemos,

\begin{align}
\omega^2_{+}  &= (2-\sqrt{2})\frac{g}{\ell} \\
\omega^2_{-} &= (2+\sqrt{2})\frac{g}{\ell} \\
\end{align}

Reemplazando queda $\phi^0_{+} = \begin{pmatrix} 1 \\ \sqrt{2} \end{pmatrix}$ y $\phi^0_{-} = \begin{pmatrix} 1 \\ -\sqrt{2} \end{pmatrix}$

La solución final queda

$$\phi = Re(A_{+} \phi^0_{+} e^{i\omega_{+}t} + A_{-}\phi^0_{-}e^{i\omega_{-}t})$$


*** Sistema geométrico ***

Se tiene un sistema con $N$ coordenadas generalizadas. La energía cinética es

$$K = \frac12\sum_{i,j} m_{ij}\dot{q}_i\dot{q}_j $$

$$V(\delta) = V(0) + \frac{1}{2}\sum_{i,j}\kappa_{ij}q_i q_j $$

La ecuación de movimiento queda entonces

$$\sum_{j}m_{ij}\ddot{q_j} = -\sum_{j}\kappa_{ij}q{j} $$

Lo que podemos reescribir como 

$$M\ddot{q} = -Kq$$

Asumiendo solución de la forma $q = e^{i\omega t}q_0$, queda un problema de determinar valores y vectores propios.

\begin{align}
\omega^2 M q_0 = K q_0 \\
(K-\omega^2M)q_0 = 0 \\
\end{align}


## Unidad 4: Ondas


Se tiene que la ecuación de ondas es
\begin{align}
\rho\frac{\partial^2y}{\partial t^2} &= T\frac{\partial^2y}{\partial x^2} \\ 
\frac{\partial^2y}{\partial t^2} &= c^2 \frac{\partial^2y}{\partial x^2} \\ 
\end{align}

Asumiendo soluciones de la forma,
$y(x,t)=e^{i\omega t}u(x)$. Llegamos a la solución

$$u(x) = A\cos{kx} + B\sin{kx}$$

Donde $k = \omega/c$

*** Dos extremos fijos ***

En el caso de los dos extremos fijos tenemos que $u(0,t) = 0$ implica $A = 0$, por lo tanto la solución queda

$$y(x,t) = B\sin{kx}$$

También se tiene que $y(L,t) = 0$ por lo tanto, $k = \frac{n\pi}{L}$

*** Un extremo en un poste ***

En este caso se tiene que $y'(L,t) = 0$, por lo que queda nuevamente $A = 0$, $k=\frac{n+\frac12}{L}\pi$. 


*** Ondas armónicas ***

Las ondas armónicas quedan descritas completamente como
$$y(x,t) = e^{ik(x-ct)}$$

Para una onda armónica definimos la {\em longitud de onda} como la distancia entre dos máximos, es decir

$$\lambda = \frac{2\pi}{k}$$

Y por otro lado definimos el periodo como el tiempo entre dos máximos,

$$T = \frac{2\pi}{\omega} = \frac{2\pi}{ck}$$

*** Modos Normales ***

En un modo normal, toda la cuerda vibra a la misma frecuencia, por lo tanto la solución queda,

$$u(x,t) = e^{i\omega t}(A\cos{kx}+B\sin{kx})$$

En una cuerda, se tiene que los extremos están fijos, por lo tanto, aplicando condiciones de borde, queda que la parte del coseno se hace nula y que el número de onda y la frecuencia siguen una cierta condición:

$$k_n = n\frac{\pi}{L} \quad \quad \quad \omega_n = n\frac{c\pi}{L}$$

La solución general queda entonces como la parte real de:

$$u(x,t) = \sum_{n=1}^{\infty}A_n e^{i\omega_n t}\sin(k_n x) $$

*** Energía ***

La energía cinética está dada por,

$$K = \int_0^L \frac12\rho\left(\frac{\partial u}{\partial t}\right)^2 dx$$

Por otro lado, la energía potencial está asociada con la tensión de la cuerta y su estiramiento:

$$V = \int_0^L T\frac12 \left(\frac{\partial u}{\partial x}\right)^2 dx $$

Finalmente, el langrageano queda

$$K-V = \int_0^L \left(\frac12\rho\left(\frac{\partial u}{\partial t}\right)^2-\frac12 T\left(\frac{\partial u}{\partial x}\right)^2\right)dx$$

El término dentro de la integral se denomina {\em densidad lagrangeana}. Pero, tenemos que $u$ es solución si se cumple que:

\begin{align}
S[u+\delta u] - S[u] &= 0 + O(\delta u^2) \\
\int dtdx\left(\rho \frac{\partial u}{\partial t} \frac{\partial \delta u}{\partial t} - T \frac{\partial u}{\partial x}\frac{\partial \delta u}{\partial x}\right) &= 0 \\
\int dtdx \left(\rho\frac{\partial^2 u}{\partial t^2} - T\frac{\partial^2 u}{\partial x^2}\right)\delta u &= 0 \\
\end{align}

Recuperamos entonces la ecuación de ondas al integrar por partes la variación de la acción.

Escribiendo la solución general como una combinación lineal de modos normales, podemos calcular la energía cinética y  la energía potencial como

$$K = \sum \frac12 \rho k_n^2 \omega_n^2 \sin^2{\omega_n t} \frac{A_n^2 + B_n^2}{2}L $$

Y la energía potencial queda

$$ V = \sum \frac12 T k_n^2 \cos^2{\omega_nt}\frac{A_n^2+B_n^2}{2}L $$

La energía mecánica es entonces conservada en el movimiento. Como el seno y coseno tienen un desfase de $\pi/2$ vemos que la energía va transformándose entre energía potencial y cinética.

*** Una cuerda colgando ***

En el caso de una cuerda colgando en presencia de un campo gravitatorio, la tensión en función de la altura $T(x) = \rho g x$. Luego, la energía cinética y potencial quedan:

\begin{align}
dK &= \frac{\rho}{2}d\ell \dot{y}^2 \\
&=\frac{\rho dx}{2} \dot{y}^2 \\
K &= \int \frac{\rho}{2} \dot{y}^2 dx \\
dV &= T(x)(d\ell - dx) \\
&= T(x)(1+\frac12 (\partial_x y)^2 - 1)dx \\
V &= \int \frac{T(x)}{2}(\partial_x y)^2 \\
\end{align}

Luego, hacemos variar la acción:

## Unidad 5: Reflexión y Refracción

Estudiaremos la potencia transmitida y la impedancia de una cuerda. Tenemos

$$ F = T\frac{\partial u}{\partial x} \quad \quad v = \frac{\partial u}{\partial t}$$

Con estas expresiones podemos determinar la potencia y la impedancia.

\begin{align}
P &= F\cdot v = - T\frac{\partial u}{\partial x}\frac{\partial u}{\partial t} \\
Z &= \frac{F}{v} = -T\frac{\frac{\partial u}{\partial x}}{\frac{\partial u}{\partial t}} 
\end{align}

En el caso de una solución de D'Alembert,

\begin{align}
P &= Tc(f'(x-ct))^2 \\
Z &= \lambda c
\end{align}

En el caso de una onda estacionaria,
\begin{align}
P &= TA^2\omega k \sin kx\cos kx \sin \omega t\cos \omega t \\
P &= \frac14 TA^2\omega k\sin 2kx\sin 2\omega t
\end{align}

Para una onda armónica que se desplaza, se obitene que
$$ P = \frac12 A^2k^2Tc $$

En el caso de una interfaz, las condiciones de borde quedan,
$$u(0^-) = u(0^+) \quad\quad T\frac{\partial u}{\partial x}_{0^+} = T\frac{\partial u}{\partial x}_{0^-}$$

Teniendo que la solución a cada lado es:
\begin{align}
u_1(x,t) &= Ae^{k_1x-\omega t} + Be^{-k_1x - \omega t} \\
u_2(x,t) &= Ce^{k_1x-\omega t}
\end{align}

De donde se deduce que,
\begin{align}
A + B &= C \\
k_1 A - k_1 B &= k_2 C
\end{align}

Por lo tanto,
\begin{align}
\frac{C}{A} &= \frac{2k_1}{k_1+k_2} \\ \\
\frac{B}{A} &= \frac{k_1-k_2}{k_1+k_2} 
\end{align}

Definimos el coeficiente de reflección $R = B/A$ y el coeficiente de transmición $T = C/A$. Además aplicando que $\omega = k_1c_1 = k_2c_2$, queda que:

\begin{align}
R &= \frac{c_2-c_1}{c_2+c_1} \\
T &= \frac{2c_2}{c_2+c_1} \\
\end{align}

### Membranas

Se tiene una superficie plana y tensa. Tenemos entoces que la energía potencial será $V = \tau(S-A)$. Por otra parte, en coordenadas cartesianas se tiene que 
$$dS = \sqrt{1+\left(\frac{\partial \zeta}{\partial x}\right)^2 + \left(\frac{\partial \zeta}{\partial y}\right)^2}dxdy$$

Así la energía potencial queda,

$$V = \tau\int_A \sqrt{1+\left(\frac{\partial \zeta}{\partial x}\right)^2 + \left(\frac{\partial \zeta}{\partial y}\right)^2} - 1 dxdy $$

Aproximando a pequeñas oscilaciones, 

$$V = \tau\int_A \frac12 \left(\left(\frac{\partial \zeta}{\partial x}\right)^2 + \left(\frac{\partial \zeta}{\partial y}\right)^2\right)dxdy$$

Por otra parte, la energía cinética está dada por

$$K = \int_A \frac12 \sigma \left(\frac{\partial \zeta}{\partial t}\right)^2 dxdy$$

Luego aplicando la ecuación de Euler-Lagrange para la densidad lagrangiana $\mathcal{L}$:

$$\frac{\partial}{\partial t}\frac{\partial \mathcal{L}}{\partial \dot{\zeta}}+\frac{\partial}{\partial x}\frac{\partial \mathcal{L}}{\partial \zeta_x}+\frac{\partial}{\partial y}\frac{\partial \mathcal{L}}{\partial \zeta_y} - \frac{\partial \mathcal{L}}{\partial \zeta} = 0$$

Luego, la ecuación queda como la ecuación de ondas en dos dimensiones:

$$\frac{\partial^2 \zeta}{\partial t^2} = c^2\left(\frac{\partial^2 \zeta}{\partial x^2} + \frac{\partial^2 \zeta}{\partial y^2}\right)$$

*** Solución ***

Asumiendo una solución de forma trigonométrica,

\begin{align}
-\omega^2 u &= c^2 \nabla^2 u \\
(\nabla^2 + k^2)u &= 0 \\
\end{align}

Esto último se conoce como la ecuación de Helmholtz


*** Otra solución: Separación de variables ***

Asumimos solución del estilo:
$$u(x,y,t) = X(x)Y(y)T(t)$$

Las condiciones iniciales para una membrana rectangular de medidas $a\times b$:

$$Y(0) = 0 \quad X(0)=0 \quad Y(b) = 0 \quad X(a) = 0$$

Reemplazamos la solución en la ecuación de ondas:

\begin{align}
X''YT+XY''T&=\frac{1}{c^2}XY\ddot{T} \\
\frac{X''}{X} + \frac{Y''}{Y} &=\frac{1}{c^2}\frac{\ddot{T}}{T} \\
\end{align}

Al asumir solución trigonométrica para $T = Ae^{i\omega t}$, con $A$ complejo:

$$\frac{X''}{X} + \frac{Y''}{Y} =-\frac{\omega^2}{c^2}$$

Definiendo $k=\frac{\omega}{c}$, y asumiendo $X$ e $Y$ como soluciones trigonométricas,

$$-k_x^2 - k_y^2 = -k^2$$

Al aplicar condiciones de borde, queda:

$$\frac{\omega_{mn}^2}{c^2} = \frac{m^2\pi^2}{a^2} + \frac{n^2\pi^2}{b^2}$$

Los modos normales son entonces:

$$u_{mn}(x,y,t) = A_{mn}\sin\left(\frac{m\pi}{a}x\right) \sin\left(\frac{n\pi}{b}y\right) \cos(\omega_{mn}t + \phi_{mn}) $$

Luego, queda
 
\begin{align}
u_0(x,y) &= \sum_{mn} A_{mn}\sin\left(\frac{m\pi}{a}x\right) \sin\left(\frac{n\pi}{b}y\right) \cos( \phi_{mn}) \\
v_0(x,y) &= \sum_{mn} -A_{mn}\omega_{mn}\sin\left(\frac{m\pi}{a}x\right) \sin\left(\frac{n\pi}{b}y\right) \sin( \phi_{mn})
\end{align}

A partir de esta expresión se puede encontrar los coeficentes de la serie de Fourier.

#### Membrana Circular:

Tenemos entonces que una membrana circular cumple:

$$\nabla^2u=\frac{1}{c^2}\frac{\partial^2u}{\partial t^2}$$

Utilizando separación de variables: $u=R(r)\Theta(\theta)e^{i\omega t}$

\begin{align}
\frac{1}{r}\frac{d}{dr} r\frac{dR}{dr} \Theta + \frac{1}{r^2}R\frac{d^2\Theta}{d\theta^2} &= -k^2 R\Theta \\
\frac{1}{rR}\frac{d}{dr} r\frac{dR}{dr} \Theta + \frac{1}{r^2}R\frac{1}{\Theta}\frac{d^2\Theta}{d\theta^2} &= -k^2 \\
\end{align}

De acá se deduce que:

$$\frac{1}{\Theta}\frac{d^2\theta}{d\theta^2} = -m^2$$

Necesitamos entonces que $m$ sea positivo. Así $\Theta(\theta) = e^{im\theta}$. Pero necesitamos que $m$ sea entero, así cuando $\theta = 2\pi$, $\Theta$ vuelve al valor que tenía en $\theta = 0$.

Si reemplazamos esto en la ecuación anterior:

$$
\frac{d^2R}{dr^2}+\frac{1}{r}\frac{dR}{dr} + (k^2-\frac{m^2}{r^2})R = 0
$$

Esto se conoce como la ecuación de Bessel que no admite solución analítica, es decir solución en base a funciones conocidas.

### Ondas de sonido en gases

Una onda de compresión y decompresión tiene la forma:
    
$$p(r,t) = P(r,t) - P_e(r)$$
    
Donde $p$ corresponde a la diferencia de presión en un punto para cada instante.

Por otro lado, la ley de hooke en este caso establece que hay un cambio de presión asociado a un cambio en el volumen:
    
$$p = -B\theta = -B\frac{\Delta V}{V}$$
    
Viendo la fuerza sobre un elemento infinitesimal de volumen:
    
$$F_x = -\frac{\partial \rho}{\partial x} dxdydz$$

Escrito de forma vectorial,

$$F = -\nabla p dxdydz$$

La última etapa es aplicar la ley de Newton,

$$-\nabla p = \rho_0 \frac{\partial^2 r}{\partial t^2}$$

Aplicando divergencia a la ecuación anterior, queda:

\begin{align}
-\nabla \cdot \nabla p &= \rho_0 \frac{\partial^2}{\partial t^2}(\nabla\cdot r) \\
 &= -\frac{\rho_0}{B} \frac{\partial^2 p}{\partial t^2} \\
\end{align}

Queda entonces, si $c^2 = \frac{B}{\rho_0}$

$$\nabla^2 p = \frac{1}{c^2}\frac{\partial^2 p}{\partial t^2}$$

*** Onda sinusoidal ***

Para el caso de una onda sinusoidal: $p(x,t) = p(x) e^{i\omega t}$

$$\nabla^2 p + k ^2 p = 0 $$

Llegamos entonces a la ecuación de Helmholtz

$$\vec{r} = \frac{1}{\omega^2 \rho_0} \nabla p$$

*** Onda plana ***

$$p(\vec{x},t) = f(\hat{n} \cdot \vec{x} - ct)$$

Satisface la ecuación de onda. Tomaremos $f$ como una función sinusoidal y $\hat{n} = \hat{x}$, es decir que,

$$p(\vec{x},t) = p_m\cos(\frac{2\pi}{\lambda}(x-ct))$$

Queda entonces que los desplazamientos son de la forma,

$$\vec{r} = -\frac{p_m}{\omega^2\rho_0}\frac{2\pi}{\lambda}\sin(\frac{2\pi}{\lambda}(x-ct))\hat{x}$$ 

*** Energía *** 

Calcularemos la energía para el caso anterior, donde $p(\hat{x},t) = p_m\cos(kx-\omega t)$

Densidad de energía cinética:

$$\frac12 \rho_0 \left(\frac{d\vec{r}}{dt}\right)^2 = \frac12 \frac{p_m^2}{\rho_0 c^2}\cos(kx-ct)$$

Densidad de energía potencial:

$$\frac12 B\theta^2 = \frac12 \frac{p^2}{B} = \frac12 \frac{p_m}{\rho_0c^2} \cos^2(kx-ct) $$
